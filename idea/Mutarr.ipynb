```python
import numpy as np
import matplotlib.pyplot as plt

def question7(p_0=0.3, p_1=0.35, p_2=0.35, nsteps=100):
    p = [p_0, p_1, p_2]
    X = np.repeat(0, nsteps)
    X[0] = 1

    for n in range(1, nsteps):
        if X[(n-1)] > 0:
            X[n] = sum(np.random.choice(range(0, len(p)), size=X[n-1], replace=True, p=p))
        else:
            X[n] = 0

    plt.plot(X)
    plt.xlabel("Generation")
    plt.ylabel("Population")
    plt.title("Branching process")
    plt.show()

# Run the code with the desired parameters
question7(p_0=0.3, p_1=0.35, p_2=0.35, nsteps=100)
```

A. Running the code 20 times, you will notice that the size of the population-level can vary between different simulations. Each simulation represents a different realization of the branching process, where random choices are made for the number of offspring at each generation based on the given probabilities. As a result, the population can exhibit different patterns and levels of growth or decline in each simulation.

B. The purpose of the line X[n] = sum(np.random.choice(range(0, len(p)), size=X[n-1], replace=True, p=p)) is to determine the population size at each generation. It uses np.random.choice to randomly choose the number of offspring for each individual in the previous generation. The sum function is used to calculate the total number of offspring, which becomes the population size for the current generation X[n].

C. Every simulation is different because of the inherent randomness introduced by the np.random.choice function. The branching process involves random choices based on the probabilities p_0, p_1, and p_2. The outcome of these random choices can vary from one simulation to another, resulting in different population sizes and growth patterns.

F. To estimate the probability of the population being extinct at 100 days with p_0=0.25, p_1=0.35, and p_2=0.4, you can modify the function question7 to return a boolean value indicating if the population became extinct by day 100:

```python
def extinction_probability(p_0=0.25, p_1=0.35, p_2=0.4, nsteps=100):
    p = [p_0, p_1, p_2]
    X = np.repeat(0, nsteps)
    X[0] = 1

    for n in range(1, nsteps):
        if X[(n-1)] > 0:
            X[n] = sum(np.random.choice(range(0, len(p)), size=X[n-1], replace=True, p=p))
        else:
            X[n] = 0

    return X[-1] == 0
```

This code runs the simulation 1000 times and counts the number of times the population became extinct by day 100. The estimated probability of extinction is calculated by dividing the count of extinct populations by the number of simulations. The result is printed as the estimated probability of extinction at 100 days.

```python
num_simulations = 1000
extinct_count = 0

for _ in range(num_simulations):
    if extinction_probability(p_0=0.25, p_1=0.35, p_2=0.4, nsteps=100):
        extinct_count += 1

probability_extinct = extinct_count / num_simulations
print("Estimated probability of extinction at 100 days:", probability_extinct)
```
